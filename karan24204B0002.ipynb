2 Program to calculate equivalent resistance in parallel

```python
R1 = float(input("Enter resistance R1 (ohms ): "))
R2 = float(input("Enter resistance R2 (ohms ): "))
Rp = (R1 * R2) / (R1+R2)
print("Equivalent resistance in series =", Rp, "ohms")
```

3 Program to calculate voltage using Ohm’s Law 

```python
I = float(input("Enter resistance I (ohms ): "))
R = float(input("Enter resistance R (ohms ): "))

V=I*R
print("Voltage =", V, "volts")
```
